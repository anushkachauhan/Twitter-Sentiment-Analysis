```python
#load lexicon of english and hindi words
def load_englishwords():
    with open('words.txt') as word_file:
        valid_words = set(word_file.read().split())

    return valid_words

def load_hinglishwords():
    with open('hinglishwords.txt') as word_file:
        valid_words = set(word_file.read().split())

    return valid_words
```

```python
import re
import nltk
import pandas as pd
from nltk.corpus import wordnet as wordnet
from nltk.corpus import sentiwordnet as swn
from nltk.corpus import stopwords
from nltk import sent_tokenize, word_tokenize
import string
from nltk.stem import PorterStemmer
ps=PorterStemmer()

from textblob import TextBlob 
#englsih sentiment
def get_tweet_sentiment(tweet): 
        ''' 
        Utility function to classify sentiment of passed tweet 
        using textblob's sentiment method 
        '''
        # create TextBlob object of passed tweet text 
        analysis = TextBlob(tweet)
        # set sentiment 
        if analysis.sentiment.polarity > 0: 
            return 'positive'
        elif analysis.sentiment.polarity == 0: 
            return 'neutral'
        else: 
            return 'negative'
        
        
        

def tweet_to_words(raw_tweet):
    
    sentence = sent_tokenize(raw_tweet)
    stemword=[]
    for s in sentence:
        s = re.sub("@\w+","",s).strip()
        s = re.sub("http\S+","",s).strip()
        letters_only = re.sub("[^a-zA-Z]", " ",s) 
        words = letters_only.lower().split()    
    
        stops = set(stopwords.words("english")) 
        meaningful_words = [w for w in words if not w in stops]
        meaningful_words =  " ".join( meaningful_words ) #a string
#         print(meaningful_words)
        
        
        words=word_tokenize(meaningful_words)
        for w in words:
#             print(stemword)
#             print("w = ",w)
            stemmed_words = ps.stem(w)
            stemword.append(stemmed_words)
#             print("Stemword=",stemword)
    stemword2=".".join(stemword)
    return stemword2
  
                        


            
        
for index in df_new.index:
    englishtext = df_new.at[index,'englishpart']
    englishtext = str(englishtext)
    englishtext=tweet_to_words(englishtext)
    #englishtext=token(englishtext)
    res=get_tweet_sentiment(englishtext)
    if res=="neutral":
        df_new.at[index,'engSent']=float(0)
    elif res=="positive":
        df_new.at[index,'engSent']=float(1)
    else:
        df_new.at[index,'engSent']=float(-1)
        
        
        
    
 
    

```

```python
from difflib import SequenceMatcher
import codecs

# Module to interact with the Hindi Senti Word Net

# Read the HSWN
class Word:

    def __init__(self,word,ppol,npol):
        self.word = word
        self.pol = ppol - npol

    def getPol(self):
        return self.pol

    def search(self,s):
        for w in self.word:
            if SequenceMatcher(a=s, b=w).ratio()>0.85:
                return True

# Contains the HSWN in memory
HSWN = []

# Searches for a word in HSWN
def searchHSWN(s):
    for word in HSWN:
        if word.search(s):
            return word.getPol()
    return 'NF'

hswn = codecs.open('HSWN1.txt','r',encoding="UTF-8").read()

hswn = hswn.split('\n')
#print(hswn)
for line in hswn:
    parts = line.split(' ')
    words = ''.join(parts[-1:]).split(',')
    #print(words)
    if len(words[0])>1:
        npol = float(''.join(parts[-2:-1]))
        ppol = float(''.join(parts[-3:-2]))
        HSWN.append(Word(words,ppol,npol))

```
