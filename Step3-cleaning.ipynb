```python
import pandas as pd
df = pd.read_csv('terror1.csv')
# df1=pd.read_csv('vivekall.csv')
```

```python
#To remove links,mentions and hashtags
for index in df.index:
    tweet_text = df.at[index, 'TweetText']
    tweet_text_split=tweet_text.split()
    #print(tweet_text_split)
    count=0
    for word in tweet_text_split:
        if word[0] == "#":
            del tweet_text_split[count]
            count=count + 1
        elif word[0]=='@':
            del tweet_text_split[count]
            count=count + 1
        elif "https" in word:
            del tweet_text_split[count]
            count=count + 1
        else:
            count=count + 1
        
    tweet_text_updated=' '.join(str(e) for e in tweet_text_split)
    #print(tweet_text_updated,"\n")
    df.at[index, 'TweetText'] =tweet_text_updated 
```

```python
#Convert emojis to text
import emoji
for index in df.index:
    tweettext = df['TweetText'][index]
    tweettext = tweettext.split()
    new_list=[]
    for word in tweettext:
        word = emoji.demojize(word)
        new_list.append(word)
        
    df.at[index, 'TweetText'] = ' '.join(str(e) for e in new_list)
    
```

```python
#total no. of english lettered tweets and non english lettered tweets
#removes non english lettered words
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True

tweet_text=[]
for index in df.index:
    tweet_text.append(df['TweetText'][index])

    
tweet2=[]
eng=0
noneng=0
          
from unidecode import unidecode
import string
punctuations=list(string.punctuation)
from nltk import sent_tokenize,word_tokenize
for tweet in tweet_text:
    word_of_sentence=[]
    sentence = sent_tokenize(tweet)
    for s in sentence:
        word = word_tokenize(s)
        
        for w in word:
            #tw = str(w)
            if w not in punctuations:
                             
                tw=isEnglish(w)
                if tw:
                    word_of_sentence.append(w)
                else:
                    word_of_sentence.append("")
    final = " ".join(e for e in word_of_sentence)
    
    tweet2.append(final)
                
                
    
#     tweet=str(tweet)
#     tweet = unidecode(tweet)
#     result=isEnglish(tweet)
#     if not result:
#         print(tweet)
#         tweet=""
#         noneng+=1
#     else:
#         eng=eng+1
#     tweet2.append(tweet)
# #eng=5207 - noneng   
# print("English Tweets=",eng)
# print("Non english = ",noneng)
```
