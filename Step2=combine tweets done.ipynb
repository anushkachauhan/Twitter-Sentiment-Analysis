```python
#convert from json to pandas
import json
tweet_files = ['newimran.json']
tweets = []
for file in tweet_files:
    with open(file, 'r') as fi:
        for line in fi.readlines():
            tweets.append(json.loads(line))
```

```python
len(tweets)
```

```python
import pandas as pd
def populate_tweet_df(tweets):
    df = pd.DataFrame()
     
    df['Screen Name'] = list(map(lambda tweet:tweet['Screen Name'],tweets))
    df['ID'] = list(map(lambda tweet:tweet['ID'],tweets))
    df['User Name'] = list(map(lambda tweet:tweet['User Name'],tweets))
    df['TweetText'] = list(map(lambda tweet: tweet['Tweet Text'], tweets))      
   
    return df
df= populate_tweet_df(tweets)
df1=populate_tweet_df(tweets)
```

```python
#to remove duplicates
df.drop_duplicates(subset ="TweetText",keep ="first", inplace = True)
# df=df.drop_duplicates()
```
